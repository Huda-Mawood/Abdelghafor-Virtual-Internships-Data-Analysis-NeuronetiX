```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hourStudy=np.array([8,9,7,10,5,11,8])
hourSleep=np.array([8,7,9,8,7,8,10])
```

```python
labels=['Saturday','Sunday','Monday','Tuesday','Wednesday','Thursday','Friday']
```

```python
fig,(ax1,ax2)=plt.subplots(1,2)
ax1.plot(labels,hourStudy,'o:r')
fig.autofmt_xdate()
ax2.plot(labels,hourSleep,'*--g')
fig.autofmt_xdate()
```
