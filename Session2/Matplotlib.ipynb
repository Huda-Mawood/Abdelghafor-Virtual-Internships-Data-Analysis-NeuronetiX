```python
#Markers
import matplotlib.pyplot as plt
import numpy as np 

x = np.array([1,2,3,4])
y = np.array([10,11,12,13])

plt.plot(x,y,'o', ms = 20, mfc = 'r')
plt.show
```

```python
#Lines
import matplotlib.pyplot as plt
import numpy as np 

x = np.array([1,2,3,4])
y = np.array([10,11,12,13])

plt.plot(x,y,lw='10')
plt.show
```

```python
#Multiple Lines
import matplotlib.pyplot as plt
import numpy as np 

x = np.array([1,2,3,4])
y = np.array([10,11,12,13])

plt.plot(x)
plt.plot(y)

plt.show
```

```python
#Lables
import matplotlib.pyplot as plt
import numpy as np 

x = np.array([1,2,3,4])
y = np.array([10,11,12,13])

plt.plot(x,y)

plt.title("Distance")
plt.xlabel("Speed")
plt.ylabel("Time")

plt.show
```

```python
#Grids
import matplotlib.pyplot as plt
import numpy as np 

x = np.array([1,2,3,4])
y = np.array([10,11,12,13])

plt.plot(x,y)

plt.title("Distance")
plt.xlabel("Speed")
plt.ylabel("Time")

plt.grid(color = 'green' , ls = '--' , lw = '0.5')

plt.show
```

```python
#Subplots
import matplotlib.pyplot as plt
import numpy as np 

#plot 1
x = np.array([1,2,3,4])
y = np.array([10,11,12,13])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title('Sales')

#plot 2 
x = np.array([2,4,6,8])
y = np.array([50,100,150,200])

plt.subplot(1,2,2)
plt.plot(x,y)
plt.title('Income')

plt.suptitle('Company')
plt.show
```

```python
#Bars
import matplotlib.pyplot as plt
import numpy as np 

x = np.array(['A','B','C','D'])
y = np.array([1,2,3,4])

plt.barh(x,y,height= 0.1)
plt.show
```
