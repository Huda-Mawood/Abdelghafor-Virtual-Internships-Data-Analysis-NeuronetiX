```python
#Line Plot
import seaborn as sns

fmri = sns.load_dataset("fmri")
sns.lineplot(x="timepoint", y="signal", data=fmri)
```

```python
#Customize Line Plot
import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset("fmri")

# customize the line plot
sns.lineplot(x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False, data=fmri)

# add labels and title
plt.xlabel("Timepoint")
plt.ylabel("Signal Intensity")
plt.title("Changes in Signal Intensity over Time")

# display the plot
plt.show()
```

```python
#Bar Plot
import seaborn as sns

titanic = sns.load_dataset("titanic")

sns.barplot(x="class", y="fare", data=titanic)
```

```python
#Customize Bar Plot
import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset("titanic")

# customize the bar plot
sns.barplot(x="class", y="fare", hue="sex", ci=None, palette="muted", data=titanic)

# add labels and title
plt.xlabel("Class")
plt.ylabel("Fare")
plt.title("Average Fare by Class and Gender on the Titanic")

# display the plot
plt.show()
```

```python
#Histograms
import seaborn as sns

iris = sns.load_dataset("iris")
sns.histplot(x="petal_length", data=iris)
```

```python
#Customize Histogram
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")

# customize the histogram
sns.histplot(data=iris, x="petal_length", bins=20, kde=True, color="green")

# add labels and title
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Lengths in Iris Flowers")

# display the plot
plt.show()
```

```python
#Dist Plot  
import numpy as np    
import seaborn as sns  
import matplotlib.pyplot as plt  

# Selecting style for boxplot with set() function  
sns.set(style="white")  

# Generate a random univariate type distribution  
ru = np.random.RandomState(10)  
d = ru.normal(size=100)  

# Plotting a simple histogram with kdeplot (Kernel Density Estimate) variation  
sns.histplot(d, kde=True, color="m")  
plot = sns.histplot(d, kde=True, color="m")  
print(plot)  
plt.show() # using show() function  
```

```python
#Heatmap Example 1
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
tips = sns.load_dataset('tips')

# Select only the numeric columns for correlation
corr = tips.select_dtypes(include='number').corr()

# Create the heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)

# Show the plot
plt.show()
```

```python
#Heatmap Example 2
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
flights = sns.load_dataset('flights')

# Pivot the data correctly
flights_pivot = flights.pivot(index='month', columns='year', values='passengers')

# Create a heatmap
sns.heatmap(flights_pivot, cmap='Blues', annot=True, fmt='d')

# Set the title and axis labels
plt.title('Passengers per month')
plt.xlabel('Year')
plt.ylabel('Month')

# Show the plot
plt.show()
```
