```python
data2=df
mean_driver_age_raw=data2["driver_age_raw"].mean()
mean_driver_age=data2["driver_age"].mean()
data2["driver_age_raw"].fillna(mean_driver_age_raw,inplace=True)
data2["driver_age"].fillna(mean_driver_age,inplace=True)
data2.fillna(method='bfill',inplace=True)
data2.info()
```

```python
data3=df
mode_driver_age_raw=data3["driver_age_raw"].mode()
mode_driver_age=data3["driver_age"].mode()
data3["driver_age_raw"].fillna(mode_driver_age_raw,inplace=True)
data3["driver_age"].fillna(mode_driver_age,inplace=True)
data3.fillna(method='bfill',inplace=True)
data3.info()
```

```python
data4=df
median_driver_age_raw=data4["driver_age_raw"].median()
median_driver_age=data4["driver_age"].median()
data4["driver_age_raw"].fillna(median_driver_age_raw,inplace=True)
data4["driver_age"].fillna(median_driver_age,inplace=True)
data4.fillna(method='bfill',inplace=True)
data4.info()
```
